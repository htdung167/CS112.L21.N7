#**BÀI 1: TẢO BIỂN**

1. **Abstraction**: Nhập n (cá thể mức 1 (ngày 0)) và k (ngày), mỗi cá thể mức i sinh ra i cá thể mức 1 và cá thể ấy tăng thêm 1 mức (i+1), tìm tổng số cá thể sau k ngày.
2. **Recognition**: Sử dụng vòng lặp for và biến tạm để đếm số lượng sau mỗi ngày.
3. **Algorithm designed**: (Lưu đồ thuật toán)
4. **Alogrithm Complexity**: O(n)


```python
n, k = map(int, input().split())
def taobien(n, k):
    n0, n1 = n,2*n
    if k == 1:
        return n1
    for i in range(k-1):
        n2 = 3*n1 - n0
        n0 = n1
        n1 = n2
    return n2%1000000007
print(taobien(n,k))
```

**ĐÁNH GIÁ**: 
  * Nhóm bọn em nhận ra dạng bài này có nét tương đồng với bài tìm số Fibonacci nên lập bảng kết quả để tìm công thức. 
  * Công thức đầu tiên tìm được là U(k) = 2*U(k-1) + U(k-2) +...+ U(0). Sau đó biến đổi một chút công thức trở nên đẹp hơn: U(k) = 3*U(k-1) - U(k-2). 
  * Đến đây nhóm bọn em đã thử cách dùng đệ quy để giải nhưng không khả thi, sau đó nhóm chuyển sang dùng vòng for và đã đúng.

#**Bài 2: BOT**

1. **Abstraction**: Chọn khoảng p,q thuộc n sao cho tổng các số a[ i ] (p<=i<=q) là lớn nhất.  
2. **Recognition**:  Thuật toán Kadane.
3. **Algorithm designed**: (Lưu đồ thuật toán)
4. **Alogrithm Complexity**: O(n)


```python
n = int(input())
lst = [int(i) for i in input().split()]
pre_sum = 0
cur_sum = 0
start = 0
_end = 0
for i in range(n):
    cur_sum = cur_sum + lst[i]
    if pre_sum < cur_sum:
        pre_sum = cur_sum
        _end = i
        cur_start = start
    if cur_sum < 0:
        cur_sum = 0
        start = i +1
print(cur_start+1,_end+1,pre_sum)
```
