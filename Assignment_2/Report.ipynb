#**BÀI 1: H_INDEX**



1. **Abstraction**:
- Tìm *res* (*res*<=*n*) lớn nhất là số các phần tử trong mảng *x*, sao cho mỗi phần tử đó có giá trị bé hơn *res*.
2. **Recognition**:
- Sử dụng hàm có sẵn "sort" và vòng lặp "for"
3. **Algorithm designed**:
- (Lưu đồ thuật toán)
4. **Alogrithm Complexity**: O(n)


image.png



5. **Programming**:

```python
n = int(input())
x = list(map(int, input().split()))

def H_Index(n, x):
    x.sort(reverse=True)
    h_index = 0
    for i in range(n):
        if x[i] >= i + 1:
            h_index += 1
        else:
            break
    return h_index

print(H_Index(n, x))
```



#**BÀI 2: KHÓA SỐ**

1. **Abstraction**: Tìm chuỗi lớn nhất chia hết cho 3 bằng cách đổi vị trí, xóa các ký tự số trong chuỗi.
2. **Pattern Recognition**:
- Tính số dư $n$ của chuỗi số khi chia cho $3$
- Tách các chữ số của chuỗi số dưới dạng 1 dictionary
#### Nếu số dư khác $0$: 
- Nếu tồn tại số có cùng số dư với chuỗi số khi chia cho $3$ thì ta sẽ bỏ một chữ số nhỏ nhất
- Nếu không tồn tại, thì bỏ hai chữ số có số dư $3-n$ khi chia cho 3 nhỏ nhất
- Sắp xếp chuỗi theo thứ tự giảm dần.
##### Ví dụ :
  1. input : 1254
>   B1    : n = 412 mod 3 = 1
>
>   B2    : Xét các chữ số trong dãy có giá trị nào chia cho $3$ dư 1  
>
>   B3    : Loại bỏ phần tử nhỏ nhất có số dư bằng với n và sắp xếp chuỗi theo thứ tự giảm dần 
>   Output: 542

  2. input : 223
>   B1    : n = 223 mod 3 = 1  
>
>   B2    : Xét các chữ số trong dãy không có chữ số nào mod 3 bằng $n$ (1)   
>
>   B3    : Xét  loại 2 chữ số nhỏ nhất có số dư chia cho 3 là  $3-n%$ là 1.
>
>   B4    : Sau đó sắp xếp chuỗi theo thứ tự giảm dần    
>
>   Output: 3

3. **Algorithm designed**:
>##### *Gọi hàm counter() đếm các chữ số trong n*
>
>`x = Counter(n)`
>
>`for i in x:`
>
>##### *Phân nhóm các chữ số trong n theo số dư khi chia cho 3*
>
>    `mod_number = [[ ] , [ i nếu i % 3 == 1] , [ i nếu i % 3 == 2 ]]`
>
>##### *Tổng các chữ số trong n*
>
>   `sum_ = sum_ + i`
>    
>##### *mod_ = phần dư của n chia cho 3*
>
> ##### *If phần dư khác 0 :*
>
>       *count_same_mod = đếm số chữ số có phần dư bằng mod_
>
>       If tồn tại count_same_mod thì:
>
>		        num_del = Xóa 1 chữ số nhỏ nhất có phần dư khi chia cho 3 bằng mod_
>
>	      Else:
>
>		        num_del = Xóa 2 số chữ số nhỏ nhất có phần dư khi chia cho 3 bằng 3 - mod_
>       Thực hiện việc xóa theo num_del
> Sắp xếp lại theo thứ tự giảm dần và in kết quả ra màn hình
4. **Alogrithm Complexity**: O(n)
5. **Programming**:
```python
from collections import Counter
x = input().strip()
count_num = Counter(x)
mod_num = []
mod_num.append([0])
mod_num.append(['1','4','7'])
mod_num.append(['2','5','8'])
mod_num = [[x for x in d if x in count_num] for d in mod_num]
sum_ = sum(int(i)*count_num[i] for i in count_num)
mod_ = sum_%3
if mod_:
    count_digit_same_mode = sum(int(count_num[i]) for i in mod_num[mod_])
    number_del = 1
    if not count_digit_same_mode:
        mod_ = 3 - mod_ 
        number_del= 2
    for i in mod_num[mod_]:
        j = min(number_del, count_num[i])
        number_del-= j
        count_num[i] -= j
for i in sorted(count_num.keys(), reverse=True):
    print(i*count_num[i], end="")
```
